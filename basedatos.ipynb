```python

try:
    conn = psycopg2.connect(
        host='172.25.30.98',
        port=5432,
        database='jbpm',
        user='jbpm',
        password='jbpm'
    )
    cursor = conn.cursor()

    query = """
    SELECT schema_name
    FROM information_schema.schemata
    WHERE schema_name NOT IN ('pg_catalog', 'information_schema')
    ORDER BY schema_name;
    """

    cursor.execute(query)
    esquemas = cursor.fetchall()

    print("üì¶ Esquemas disponibles:")
    for esquema in esquemas:
        print(f"  - {esquema[0]}")

    cursor.close()
    conn.close()

except Exception as e:
    print("‚ùå Error al consultar esquemas:", e)
```
